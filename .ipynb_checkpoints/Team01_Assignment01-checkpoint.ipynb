Import python packages

```python
import numpy as np                        # ndarrys for gridded data
import pandas as pd                       # DataFrames for tabular data
import os                                 # set working directory, run executables
import matplotlib.pyplot as plt           # for plotting
import geostatspy.GSLIB as GSLIB          # GSLIB utilies, visualization and wrapper
import geostatspy.geostats as geostats    # GSLIB methods convert to Python        
```

Load in data to a pandas DataFrame and numpy ndarray

```python
df = pd.read_csv('1_sample_data.csv')
seismic = np.loadtxt('1_AI.csv',delimiter=',')
```

Configure plotting parameters and then plot the spatial sample points of the facies, porosity, permeability, and acoustic impedence.

```python
xmin = 0.0; xmax = 1000.0               # range of x values
ymin = 0.0; ymax = 1000.0               # range of y values
pormin = 0.0; pormax = 0.3;             # range of porosity values
cmap = plt.cm.plasma                    # color map
permmin = 0.01; permmax = 3000;
AImin = 1000.0; AImax = 8000
Fmin = 0; Fmax = 1
```

```python
plt.subplot(221)
GSLIB.locmap_st(df,'X','Y','Facies',xmin,xmax,ymin,ymax,Fmin,Fmax,'Well Data - Facies','X(m)','Y(m)','Sand (1), Shale (0)',plt.cm.inferno)

plt.subplot(222)
GSLIB.locmap_st(df,'X','Y','Porosity',xmin,xmax,ymin,ymax,0.06,0.24,'Well Data - Porosity','X(m)','Y(m)','Porosity (fraction)',cmap)

plt.subplot(223)
GSLIB.locmap_st(df,'X','Y','Perm',xmin,xmax,ymin,ymax,permmin,permmax,'Well Data - Permeability','X(m)','Y(m)','Permeaiblity (mD)',cmap)

plt.subplot(224)
GSLIB.locmap_st(df,'X','Y','AI',xmin,xmax,ymin,ymax,AImin,AImax,'Well Data - Acoustic Impedance','X(m)','Y(m)','Acoustic Impedance (kg/m2s*10^6)',cmap)

plt.subplots_adjust(left=0.0, bottom=0.0, right=3.0, top=3.5, wspace=0.1, hspace=0.2)
plt.savefig('All_location_maps.tif',dpi=600,bbox_inches="tight")
plt.show()
```

Single histogram of all porosity sample data.

```python
# univariate histograms of all samples
n_bins = 15

fig, ax1 = plt.subplots(2,2, figsize=(20,15))

# Facies
ax1[0,0].hist(df['Facies'], bins=n_bins, alpha=0.5, label='All Data')
ax1[0,0].set_title('Facies Histogram')
ax1[0,0].set_xlabel('Facies')
ax1[0,0].set_ylabel('Frequency')
ax1[0,0].legend()

# Porosity
ax1[0,1].hist(df['Porosity'], bins=n_bins, alpha=0.5, label='All Data')
ax1[0,1].set_title('Porosity Histogram')
ax1[0,1].set_xlabel('Porosity')
ax1[0,1].set_ylabel('Frequency')
ax1[0,1].legend()

# Permeability
ax1[1,0].hist(df['Perm'], bins=n_bins, alpha=0.5, label='All Data')
ax1[1,0].set_title('Permeability Histogram')
ax1[1,0].set_xlabel('Permeability')
ax1[1,0].set_ylabel('Frequency')
ax1[1,0].legend()

# Acoustic Impedence
ax1[1,1].hist(df['AI'], bins=n_bins, alpha=0.5, label='All Data')
ax1[1,1].set_title('Acoustic Impedence Histogram')
ax1[1,1].set_xlabel('Porosity')
ax1[1,1].set_ylabel('Acoustic Impedence')
ax1[1,1].legend()

plt.show()
```

Splitting the data by facies, and then plotting the porosity histograms for each facies.

```python
# univariate histograms split by facies
facies_0 = df[df['Facies']==0]
facies_1 = df[df['Facies']==1]

n_bins = 15

fig, ax2 = plt.subplots(2,2, figsize=(20,15))

# Facies
ax2[0,0].hist(facies_0['Facies'], bins=n_bins, alpha=0.4, color='b', label='Facies_0')
ax2[0,0].hist(facies_1['Facies'], bins=n_bins, alpha=0.4, color='r', label='Facies_1')
ax2[0,0].set_title('Facies Histogram Split By Facies')
ax2[0,0].set_ylabel('Count')
ax2[0,0].set_xlabel('Facies')
ax2[0,0].legend()

# Porosity
ax2[0,1].hist(facies_0['Porosity'], bins=n_bins, alpha=0.4, density=True, color='b', label='Facies_0')
ax2[0,1].hist(facies_1['Porosity'], bins=n_bins, alpha=0.4, density=True, color='r', label='Facies_1')
ax2[0,1].set_title('Porosity Histogram Split By Facies')
ax2[0,1].set_ylabel('Density')
ax2[0,1].set_xlabel('Porosity')
ax2[0,1].legend()

# Permeability
ax2[1,0].hist(facies_0['Perm'], bins=n_bins, alpha=0.4, density=True, color='b', label='Facies_0')
ax2[1,0].hist(facies_1['Perm'], bins=n_bins, alpha=0.4, density=True, color='r', label='Facies_1')
ax2[1,0].set_title('Permeability Histogram Split By Facies')
ax2[1,0].set_ylabel('Density')
ax2[1,0].set_xlabel('Permeability')
ax2[1,0].legend()

# Acoustic Impedence
ax2[1,1].hist(facies_0['AI'], bins=n_bins, alpha=0.4, density=True, color='b', label='Facies_0')
ax2[1,1].hist(facies_1['AI'], bins=n_bins, alpha=0.4, density=True, color='r', label='Facies_1')
ax2[1,1].set_title('Acoustic Impedence Histogram Split By Facies')
ax2[1,1].set_ylabel('Density')
ax2[1,1].set_xlabel('Acoustic Impedence')
ax2[1,1].legend()

plt.show()
```
